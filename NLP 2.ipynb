```python
import nltk
nltk.download('punkt')
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import word_tokenize
stop=stopwords.words('english')

t=["Thor Hathodawala is looking for a job"]
#t="".join(t)
#wt=word_tokenize(t)
#t=[w for w in wt if w not in stop]

cv=CountVectorizer(ngram_range=(1,2))
cv.fit(t)
cv.vocabulary_
#cv.get_feature_names_out()
```

```python
from nltk.stem import PorterStemmer

corpus = [
    "Thor ate pizza",
    "Loki is tall",
    "Loki is eating pizza"
]
tokens=[]

for t in corpus:
  wt=word_tokenize(t)
  ps=PorterStemmer()
  stem=[ps.stem(w) for w in wt if w not in stop]
  stem=" ".join(stem)
  tokens.append(stem)

#print(tokens)

cv=CountVectorizer(ngram_range=(1,2))
cv.fit(tokens)
cv.vocabulary_
#cv.get_feature_names_out()
```

```python
import pandas as pd
df=pd.read_json("news_dataset.json")
df.head()
```

```python
df.category.value_counts()
```

```python
df_sc=df[df['category']=='SCIENCE']
min_samples=df_sc.shape[0]

df_bus = df[df.category=="BUSINESS"].sample(min_samples, random_state=2022)
df_sp = df[df.category=="SPORTS"].sample(min_samples, random_state=2022)
df_crm = df[df.category=="CRIME"].sample(min_samples, random_state=2022)

df_balanced=pd.concat([df_bus,df_sp,df_crm,df_sc],axis=0)
df_balanced.category.value_counts()
```

```python
df_balanced.head()
```

```python
df_balanced.category.replace({'BUSINESS': 0,
    'SPORTS': 1,
    'CRIME': 2,
    'SCIENCE': 3},inplace=True)
df_balanced.head()
```

```python
cv=CountVectorizer(ngram_range=(1,1))
cv_matrix=cv.fit_transform(df_balanced.text)
cv_matrix=cv_matrix.toarray()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(cv_matrix,df_balanced.category,test_size=0.2,random_state=2022, stratify = df_balanced.category)
```

```python
from sklearn.naive_bayes import MultinomialNB

model=MultinomialNB()
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report

cr=classification_report(Y_test,y_pred)
print(cr)
```

```python
url = 'https://zenodo.org/records/3355823/files/ecommerceDataset.csv'
import pandas as pd

import csv, urllib.request


response = urllib.request.urlopen(url)
lines = [l.decode('utf-8') for l in response.readlines()]
cr = csv.reader(lines)

df=pd.DataFrame(cr,columns=['Label','Text'])
df.head()
#for row in cr:
#    print(row)
```

```python
df['Label'].value_counts()
```

```python
print(df.shape)
```

```python
min_samples=2000

df_h = df[df.Label=="Household"].sample(min_samples, random_state=2022)
df_b = df[df.Label=="Books"].sample(min_samples, random_state=2022)
df_e = df[df.Label=="Electronics"].sample(min_samples, random_state=2022)
df_c = df[df.Label=="Clothing & Accessories"].sample(min_samples, random_state=2022)

df_balanced=pd.concat([df_h,df_b,df_c,df_e],axis=0)
df_balanced.Label.value_counts()
```

```python
df_balanced.Label.replace({'Household': 0,
    'Books': 1,
    'Electronics': 2,
    'Clothing & Accessories': 3},inplace=True)
df_balanced.head()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(df_balanced.Text,df_balanced.Label,test_size=0.2,random_state=2022, stratify = df_balanced.Label)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
from sklearn.feature_extraction.text import TfidfVectorizer

#1. create a pipeline object
clf = Pipeline([
     ('vectorizer_tfidf',TfidfVectorizer()),
     ('KNN', KNeighborsClassifier())
])

#2. fit with X_train and y_train
clf.fit(X_train, y_train)


#3. get the predictions for X_test and store it in y_pred
y_pred = clf.predict(X_test)


#4. print the classfication report
print(classification_report(y_test, y_pred))
```
