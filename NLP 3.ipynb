```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('SMSSpamCollection',names=["label", "message"], sep='\t')
df
```

```python
df['label'].value_counts()
```

```python
df['Spam']=df.label
```

```python
df.Spam.replace({'ham':0,'spam':1},inplace=True)
df.head()
```

```python

from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer()
cv_matrix=cv.fit_transform(df.message)
cv_matrix=cv_matrix.toarray()

cv_matrix
```

```python
cv.get_feature_names_out()
```

```python
cv.get_feature_names_out().shape
```

```python
cv.vocabulary_
```

```python
cv.get_feature_names_out()[2050]
```

```python
cv_matrix.shape
```

```python
cv_matrix[0]
```

```python
np.where(cv_matrix[0]!=0)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(cv_matrix,df.Spam,test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB

model=MultinomialNB()
model.fit(X_train,Y_train)
```

```python
model.score(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report

cr=classification_report(Y_test,y_pred)
print(cr)
```
