```python
from gensim.models import Word2Vec

tokenized_sentences = [['Hello','This','is','python','training','by','faculty','hen'],
             ['Hello','This','is','Java','training','by','faculty'],
             ['Hello','This','is','Data Science','training','by','Unfold','Data','Science','hen'],
             ['Hello','This','is','programming','training','']]

model=Word2Vec(tokenized_sentences,min_count=2,vector_size=200)
print(model)
```

```python
model.wv.index_to_key  # count 2 and above items
```

```python
model.wv.key_to_index
```

```python
model.wv['is']  #vector
```

```python
model.wv.most_similar("faculty")
```
