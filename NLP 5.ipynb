```python
# Making a GET request
r = requests.get('https://www.worldometers.info/coronavirus/countries-where-coronavirus-has-spread/#google_vignette/')

# Parsing the HTML
soup = BeautifulSoup(r.text, 'html.parser')
#print(soup.prettify())

tab=soup.find('table',{'id':"table3"})
#print(tab)

title=[]
headers=tab.find_all('th')
for i in headers:
  print(i.text)
  title.append(i.text)
```

```python
import pandas as pd

df=pd.DataFrame(columns=title)
df
```

```python
country=[]
cases=[]
deaths=[]
region=[]

row=tab.find_all('tr')

for i in row[1:]:
  data=i.find_all('td')
  #print(data)
  each=[j.text for j in data]
  l=df.shape[0]
  df.loc[l]=each

df.head()
```

```python
from tabulate import tabulate
print(tabulate(df, headers = 'keys', tablefmt = 'psql'))
```

```python
import requests
from  bs4 import BeautifulSoup

url='https://www.iplt20.com/auction'
r=requests.get(url)

soup=BeautifulSoup(r.text,'lxml')
tab=soup.find_all('div',{'class':"auction-grid-view mt-3"})

for i in tab:
  #print(i)
  teamlist=i.find_all('div',{'class':'agv-team-name'})
  Teams=[j.text for j in teamlist]
  #-----
  Fundlist=i.find_all('span',{'class':'fr-fund'})
  Funds=[j.text for j in Fundlist]
  #print(Funds)
  break

x=0
a=[]
b=[]
c=[]
for i in range(0,10):
  a.append(Funds[x])
  b.append(Funds[x+1])
  c.append(Funds[x+2])
  x=x+3

print(a,'\n',b,'\n',c)
```
